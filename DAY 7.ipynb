# 2. Perform Poisson distribution

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
sns.displot(random.poisson(lam=10,size=200))
plt.show()
```

# 3. Perform Normal, Exponential distributions

```python
exp= np.random.exponential(2,200)
count,bins,ignored=plt.hist(exp,10)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

mu,sigma=0.8,0.5
s=np.random.normal(mu,sigma,10000)

count,bins,ignored=plt.hist(s,1000)

#distribution curve:
plt.plot(bins,1/sigma*np.sqrt(2*np.pi)*np.exp(-(bins-mu)**2/(2*sigma**2)))

plt.show()
```
